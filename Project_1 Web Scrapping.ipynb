```python
from bs4 import BeautifulSoup
import requests as r
import pandas as pd
```

# Using BeautifulSoup to extract table to DataFrame 

```python
url = "http://www.espn.com/nba/salaries"
data = r.get(url).text
soup = BeautifulSoup(data, 'html.parser')
```

```python
table = soup.find_all('table')
len(table)
```

```python
table = table[0]
heads = table.find_all('td')[:4]
columns = [head.text.strip() for head in heads]
nba_df = pd.DataFrame(columns =columns)
for index, row in enumerate(table):
    cols = row.find_all('td')
    add = []
    if 'RK' not in cols:
        rk = cols[0].text
        name = cols[1].text.strip()
        team = cols[2].text.strip()
        salary = cols[3].text
        add = [rk,name,team,salary]
        for i, val in zip(columns, add):
            nba_df.loc[index, i] = val
row_del = nba_df[nba_df.RK == 'RK']
nba_df = nba_df.drop(index=row_del.index)
nba_df.index =  [i for i in range(len(nba_df))]
nba_df
```
